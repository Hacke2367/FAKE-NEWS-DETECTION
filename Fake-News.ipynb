```python
import nltk
import string
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer= WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'https\S+', '', text)
    text = re.sub(r'\d+', '', text)
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word not in string.punctuation]
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return " ".join(tokens)
```

```python
data['clean_text'] = data['text'].apply(clean_text)
```

```python
data['full_text'] = data['title'] + " " + data['text']
data['fully_clean'] = data['full_text'].apply(clean_text)
data['clean_title'] = data['title'].apply(clean_text)
```

```python
from nltk import ngrams, FreqDist

def get_top_biagrams(text_series, n=10):
    bigrams = []
    for text in text_series:
        words = text.split()
        bigrams.extend(list(ngrams(words,2)))
    return FreqDist(bigrams).most_common(n)


real_bigrams = get_top_biagrams(data[data['label']== 1]['fully_clean'])
print(f"\n Real News top bigram", real_bigrams)

fake_bigrams = get_top_biagrams(data[data['label'] == 0]['fully_clean'])
print(f"\n Fake News Top biagram ", fake_bigrams)
        
```

```python
from nltk import ngrams, FreqDist

# Function to get top bigrams
def get_top_bigrams(text_series, n=10):
    bigrams = []
    for text in text_series:
        words = text.split()
        bigrams.extend(list(ngrams(words, 2)))
    return FreqDist(bigrams).most_common(n)

# Real News Top Bigrams
real_bigrams = get_top_bigrams(data[data['label']==0]['fully_clean'])
print("\nReal News Top Bigrams:", real_bigrams)

# Fake News Top Bigrams
fake_bigrams = get_top_bigrams(data[data['label']==1]['fully_clean'])
print("\nFake News Top Bigrams:", fake_bigrams)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf = TfidfVectorizer(max_features=5000)
X= tfidf.fit_transform(data['fully_clean']).toarray()
y = data['label']

print('tfidf', X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,data['label'], test_size=0.2, random_state=42)
```

```python

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Step 3: Predict Function for New Query
def predict_news(news_text):
    cleaned = clean_text(news_text)
    vector = tfidf.transform([cleaned]).toarray()
    result = model.predict(vector)[0]

    if result == 1:
        return "✅ Real News"
    else:
        return "❌ Fake News"
```
